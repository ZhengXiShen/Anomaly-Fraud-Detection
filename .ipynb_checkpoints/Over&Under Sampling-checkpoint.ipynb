### Detect special values lik 'NaN', '-99', '-98', '0'. And keep them as a category since they usually contain some useful system information.

```python
# '-99', '-98'
mask = (train == -99) | (train == -98)

columns_with_sepcial_val = df.columns[mask.any()]

print(columns_with_sepcial_val)
```

```python
# '0'
columns_with_sepcial_val = train.columns[(train == 0).any()]

print(columns_with_sepcial_val)
```

```python
# 'NaN'
columns_with_nan = train.columns[train.isnull().any()]

print(columns_with_nan)
```

```python
# Create a full list of all the column contain special values

mask = (train == -99) | (train == -98) | train.isnull() | (train == 0)

columns_with_sepcial_val = train.columns[mask.any()]

print(columns_with_sepcial_val)
```

```python
# Select only the numerical columns to do binning
numerical_columns = train[columns_with_sepcial_val].select_dtypes(include=['float']).columns.tolist()

print(numerical_columns)
```

```python
# Write a function to do binning while keep the special values

def binWithSwpcial(column, data):
  data[column + '_D'] = pd.qcut(data[column], 10, duplicates = 'drop').cat.add_categories('NoData')
  data[column + '_D'] = data[column + '_D'].fillna('NoData').astype('str')
  data.loc[data[column]==-99, column + '_D'] = -99
  data.loc[data[column]==-98, column + '_D'] = -98
  data.loc[data[column]==0, column + '_D'] = 0
```

```python
# Apply the binning function to numerical columns selected above

for c in numerical_columns:
  if c != 'TD055':
    binWithSwpcial(c, train)
```

### Apply the binning and other engineering we did with train data to testdata with the same criteria

```python
# New functiopn to apply the binning of train data to test data
def applyBinning(column, train_data, test_data):
    # Create bins based on the training data
    train_data[column + '_D'], bins = pd.qcut(train_data[column], 10, duplicates = 'drop', retbins=True)
    train_data[column + '_D'] = train_data[column + '_D'].cat.add_categories('NoData')
    train_data[column + '_D'] = train_data[column + '_D'].fillna('NoData').astype('str')
    train_data.loc[train_data[column]==-99, column + '_D'] = '-99'
    train_data.loc[train_data[column]==-98, column + '_D'] = '-98'
    train_data.loc[train_data[column]==0, column + '_D'] = '0'

    # Apply the same bins to the test data
    test_data[column + '_D'] = pd.cut(test_data[column], bins, include_lowest=True)
    test_data[column + '_D'] = test_data[column + '_D'].cat.add_categories('NoData')
    test_data[column + '_D'] = test_data[column + '_D'].fillna('NoData').astype('str')
    test_data.loc[test_data[column]==-99, column + '_D'] = '-99'
    test_data.loc[test_data[column]==-98, column + '_D'] = '-98'
    test_data.loc[test_data[column]==0, column + '_D'] = '0'


for c in numerical_columns:
  if c != 'TD055':
    applyBinning(c, train, test)
```

### Run a small smaple dataset first

```python
# Split the smaple sample
train_smpl = train_woe.sample(frac=0.1, random_state=1)
test_smpl = test_woe.sample(frac=0.1, random_state=1)

# Trandfer the data to h2o's frame
train_hex = h2o.H2OFrame(train_smpl)
test_hex = h2o.H2OFrame(test_smpl)
```

```python
# Small sample random forest model training
rf_v1 = H2ORandomForestEstimator(
        model_id = 'rf_v1',
        ntrees = 300,
        nfolds=10,
        min_rows=100,
        seed=1234)
rf_v1.train(predictors,target,training_frame=train_hex)
```

```python
# Model prediction and comparation to the test y
predictions = rf_v1.predict(test_hex)
test_scores = test_hex['loan_default'].cbind(predictions).as_data_frame()
test_scores.head()
```

```python
# Generate a function to draw ROC curve and precision - recall curve
def ROC_AUC(my_result,df,target):
    from sklearn.metrics import roc_curve,auc
    from sklearn.metrics import average_precision_score
    from sklearn.metrics import precision_recall_curve
    import matplotlib.pyplot as plt

    # ROC
    y_actual = df[target].as_data_frame()
    y_pred = my_result.predict(df).as_data_frame()
    fpr = list()
    tpr = list()
    roc_auc = list()
    fpr,tpr,_ = roc_curve(y_actual,y_pred)
    roc_auc = auc(fpr,tpr)

    # Precision-Recall
    average_precision = average_precision_score(y_actual,y_pred)

    # plotting
    plt.figure(figsize=(10,4))

    # ROC
    plt.subplot(1,2,1)
    plt.plot(fpr,tpr,color='darkorange',lw=2,label='ROC curve (aare=%0.2f)' % roc_auc)
    plt.plot([0,1],[0,1],color='navy',lw=3,linestyle='--')
    plt.xlim([0.0,1.0])
    plt.ylim([0.0,1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic: AUC={0:0.4f}'.format(roc_auc))
    plt.legend(loc='lower right')

    # Precision-Recall
    plt.subplot(1,2,2)
    precision,recall,_ = precision_recall_curve(y_actual,y_pred)
    plt.step(recall,precision,color='b',alpha=0.2,where='post')
    plt.fill_between(recall,precision,step='post',alpha=0.2,color='b')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0,1.05])
    plt.xlim([0.0,1.0])
    plt.title('Precision-Recall curve: PR={0:0.4f}'.format(average_precision))
    plt.show()
```

```python
ROC_AUC(rf_v1,test_hex,'loan_default')
```

### Tuning Hyperparameter to improve the model performance

```python
# Define the hyperparameters to search over

from h2o.grid import H2OGridSearch

hyper_params = {'ntrees': [200, 300, 400],
                'max_depth': [10, 15, 20],
                'min_rows': [50, 100, 200]}

rf_tune = H2ORandomForestEstimator(
    nfolds = 5,
    seed = 1234
    )

grid = H2OGridSearch(model=rf_tune, hyper_params=hyper_params)
grid.train(predictors, target, training_frame = train_hex)

best_model = grid.get_grid()[0]
best_params = best_model.params

# Evaluate the best model on the test data
performance = best_model.model_performance(test_hex)

# Print the best hyperparameters and model performance
print("Best Hyperparameters:", best_params)
print("Model Performance:", performance)
```

### The Best hyperparameters from tunning are:
ntrees: 300  

max_depth: 15  

min_rows: 50

```python
# Using the best hyperparameters
rf_tune = H2ORandomForestEstimator(
    model_id = 'rf_tune',
    ntrees = 300,
    nfolds = 5,
    min_rows = 50,
    max_depth = 15,
    seed = 1234
)

rf_tune.train(predictors, target, training_frame = train_hex)
```

```python
# Draw the ROC curve and precision - recall curve to view the model performance
ROC_AUC(rf_tune, test_hex, 'loan_default')
```

## Under Sampling

### Since there are more non-default(0) values, so we should try undersampling them to see if can improve the performace of the random Forest model

```python
from imblearn.under_sampling import (TomekLinks,
                    NeighbourhoodCleaningRule,
                    NearMiss)
```

```python
y_train = train_woe[target]
X_train = train_woe.drop(target, axis = 1)
```

```python
# TomekLinks
sampler = TomekLinks()
X_rs_tl, y_rs_tl = sampler.fit_resample(X_train, y_train)
```

```python
train_tl = pd.concat([X_rs_tl, y_rs_tl], axis=1)
tl_hex = h2o.H2OFrame(train_tl)
```

```python
rf_tl = H2ORandomForestEstimator(
    model_id = 'rf_tl',
    ntrees = 300,
    max_depth = 15,
    min_rows = 50,
    nfolds = 5,
    seed = 1234
)

rf_tl.train(predictors, target, training_frame = tl_hex)
```

```python
ROC_AUC(rf_tl, test_hex, 'loan_default')
```

```python
# NeighbourhoodCleaningRule
sampler = NeighbourhoodCleaningRule()
X_rs_ncr, y_rs_ncr = sampler.fit_resample(X_train, y_train)
```

```python
train_ncr = pd.concat([X_rs_ncr, y_rs_ncr], axis=1)
ncr_hex = h2o.H2OFrame(train_ncr)
```

```python
rf_ncr = H2ORandomForestEstimator(
    model_id = 'rf_ncr',
    ntrees = 300,
    max_depth = 15,
    min_rows = 50,
    nfolds = 5,
    seed = 1234
)

rf_ncr.train(predictors, target, training_frame = ncr_hex)
```

```python
ROC_AUC(rf_ncr, test_hex, 'loan_default')
```

```python
# NearMiss
sampler = NearMiss()
X_rs_nm, y_rs_nm = sampler.fit_resample(X_train, y_train)
```

```python
train_nm = pd.concat([X_rs_nm, y_rs_nm], axis=1)
nm_hex = h2o.H2OFrame(train_nm)
```

```python
rf_nm = H2ORandomForestEstimator(
    model_id = 'rf_nm',
    ntrees = 300,
    max_depth = 15,
    min_rows = 50,
    nfolds = 5,
    seed = 1234
)

rf_nm.train(predictors, target, training_frame = nm_hex)
```

```python
ROC_AUC(rf_nm, test_hex, 'loan_default')
```

# Under Sampling result:
Among the three we used, the one with the best performance is 'TomekLinks'. It brought the result of: AUC = 0.6462 and PR = 0.2874.    
Tomeklink: It works by identifying pairs of instances from different classes that are the closest to each other, called Tomek pairs, and removing the majority class instance from each pair. This process helps improve the performance of classifiers by increasing the distance between classes and making the decision boundary clearer. TomekLinks is particularly useful when dealing with overlapping regions between classes, as it focuses on eliminating instances that contribute to the ambiguity of the data, thereby enhancing the model's accuracy and robustness.  

While 'NearMiss' brought a significant decrease in the model performance, even worse than the original random forest.

## Over Sampling

```python
from imblearn.over_sampling import (RandomOverSampler,
                   SMOTE,
                   ADASYN)
```

```python
y0_cnt_over = train_woe.shape[0] - train_woe['loan_default'].sum()
y0_cnt_over
```

```python
y1_cnt_over = (y0_cnt_over / 2).astype(int)
y1_cnt_over
```

```python
# RandomOverSampler

sampler = RandomOverSampler(sampling_strategy={1: y1_cnt_over, 0: y0_cnt_over})
X_rs_ros, y_rs_ros = sampler.fit_resample(X_train, y_train)
```

```python
train_ros = pd.concat([X_rs_ros, y_rs_ros], axis=1)
ros_hex = h2o.H2OFrame(train_ros)
```

```python
rf_ros = H2ORandomForestEstimator(
    model_id = 'rf_ros',
    ntrees = 300,
    max_depth = 15,
    min_rows = 50,
    nfolds = 5,
    seed = 1234
)

rf_ros.train(predictors, target, training_frame = ros_hex)
```

```python
ROC_AUC(rf_ros, test_hex, 'loan_default')
```

```python
# SMOTE

sampler = SMOTE(sampling_strategy={1: y1_cnt_over, 0: y0_cnt_over})
X_rs_smote, y_rs_smote = sampler.fit_resample(X_train, y_train)
```

```python
train_smote = pd.concat([X_rs_smote, y_rs_smote], axis=1)
smote_hex = h2o.H2OFrame(train_smote)
```

```python
rf_smote = H2ORandomForestEstimator(
    model_id = 'rf_smote',
    ntrees = 300,
    max_depth = 15,
    min_rows = 50,
    nfolds = 5,
    seed = 1234
)

rf_smote.train(predictors, target, training_frame = smote_hex)
```

```python
ROC_AUC(rf_smote, test_hex, 'loan_default')
```

```python
# ADASYN

sampler = ADASYN(sampling_strategy={1: y1_cnt_over, 0: y0_cnt_over})
X_rs_adasyn, y_rs_adasyn = sampler.fit_resample(X_train, y_train)
```

```python
train_adasyn = pd.concat([X_rs_adasyn, y_rs_adasyn], axis=1)
adasyn_hex = h2o.H2OFrame(train_adasyn)
```

```python
rf_adasyn = H2ORandomForestEstimator(
    model_id = 'rf_adasyn',
    ntrees = 300,
    max_depth = 15,
    min_rows = 50,
    nfolds = 5,
    seed = 1234
)

rf_adasyn.train(predictors, target, training_frame = adasyn_hex)
```

```python
ROC_AUC(rf_adasyn, test_hex, 'loan_default')
```

### Over Sampling result:
The over sampling with the best performance is the 'ADASYN', with AUC = 0.6456 and PR 0.2876.  
It focuses on those instances that are difficult to learn by the classifier. ADASYN identifies the minority class instances with fewer neighboring majority class instances and creates synthetic samples for these instances with higher densities of neighboring majority class points. This adaptively boosts the representation of the minority class, addressing class imbalance effectively. ADASYN helps to mitigate the impact of data skew on model performance, enhancing the classifier's ability to learn from the minority class and improving overall predictive accuracy in imbalanced datasets.  

There is no significant drop of model performance with all three over sampling method we use.
