## Load Dataset

```python
import pandas as pd

# Read the data into a DataFrame
df = pd.read_csv('.../DOM_hourly.csv')

# Now you can work with the DataFrame
print(df.shape)  
```

```python
print(df.tail())
```

```python
# Convert date from hourly to daily

from datetime import datetime

df['Datetime'] = pd.to_datetime(df['Datetime'])

# Use 7am data to repersent whole day
df = df[(df['Datetime'].dt.hour == 7)]

df = df.sort_values(by='Datetime')


print(df.head(10))
print(df.shape)
```

## Clean Data

```python
# Drop None value

df = df.dropna()

print(df.shape)
```

## Fit models to detect anomalies

## Simple Moving Average(SMA)

```python
# Create a dupulicated df for SMA 
df_sma = df

# Calculate the moving average of the previous 5 values for each day
df_sma['SMA'] = df_sma.loc[:, 'DOM_MW'].rolling(window = 5).mean()

df_sma['SMA_diff'] = df_sma['DOM_MW'] - df_sma['SMA']

df_sma.head()
```

```python
# Plot the actual demand and SMA prodicted demand to see the accurancy

df_sma[['DOM_MW','SMA']].plot()
```

```python
# Draw a histogram to show the differnec of actual and predited values to evaluate the tolerance range
df_sma['SMA_diff'].hist()
```

```python
# Calculate the upper and lower bounds of estimated demand based on the tolerance range above

df_sma['upper'] = df_sma['SMA'] + 2000
df_sma['lower'] = df_sma['SMA'] - 2000

df_sma.head(10)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

# Reveal the outliers by the tolerance band we got above

def plot_it():
    plt.plot(np.arange(df_sma.shape[0]),df_sma['DOM_MW'],'go',markersize=2,label='Actual')
    plt.fill_between(
       np.arange(df_sma.shape[0]), df_sma['lower'], df_sma['upper'], alpha=0.5, color="r",
       label="Predicted interval")
    plt.xlabel("Ordered samples.")
    plt.ylabel("Values and prediction intervals.")
    plt.show()

plt.figure(figsize=(12, 8))    
plot_it()
```

### Conclusion
#### By calculating the upper and lower bounds of the estimated demand based on the tolerance range derived from the Simple Moving Average (SMA) model, we can identify anomalies in the energy usage time series data. The plotted graph displays the actual energy usage values(green dots) and the predicted interval(orange shaded region), visually highlighting any outliers that fall outside the tolerance band(green dots that do not fall on the orange area). From a technical point of view, observations outside the interval suggested by the Simple Moving Average (SMA) model can be detected as anomalies because they significantly deviate from the predicted values based on the historical moving average, indicating potential abnormal fluctuations or unexpected changes in energy usage. From a common sense perspective, these observations outside the predicted interval can be considered anomalies as they represent energy consumption that falls outside the normal range expected based on historical averages, suggesting unusual or atypical energy usage patterns that may require further scrutiny or investigation.

## Exponential Smoothing

```python
from statsmodels.tsa.api import SimpleExpSmoothing

# Duplicate df for EMA
df_ema = df

# Fit the EMA model on the 'training' data with parameters
EMAfit = SimpleExpSmoothing(df['DOM_MW']).fit(smoothing_level = 0.2, optimized = False)

# Use fitted model to predict electricity demand
df_ema['EMA'] = EMAfit.predict(start = 0)

# Calcluate the difference between prediction and real value
df_ema['diff'] = df_ema['DOM_MW'] - df_ema['EMA']

df_ema[['DOM_MW', 'EMA']].plot()
```

```python
# Histgrame of the differneces
df_ema['diff'].hist()
```

```python
import numpy as np

# Create the tolerance range base on the difference histgram
df_ema['upper'] = df_ema['EMA'] + 3000
df_ema['lower'] = df_ema['EMA'] - 3000

# Plot to see whether all the electricity demand data falls into the tolerance range, to find potential anomalies
def plot_it():
    plt.plot(np.arange(df_ema.shape[0]),df_ema['DOM_MW'],'go',markersize=2,label='Actual')
    plt.fill_between(
       np.arange(df_ema.shape[0]), df_ema['lower'], df_ema['upper'], alpha=0.5, color="r",
       label="Predicted interval")
    plt.xlabel("Ordered samples.")
    plt.ylabel("Values and prediction intervals.")
    plt.show()

plt.figure(figsize=(12, 8))     
plot_it()
```

### Conclusion
#### The Exponential Smoothing model is utilized to identify anomalies in the energy usage time series data. This model forecasts future values based on the historical data and assigns higher weights to recent observations. By fitting the Exponential Smoothing model to the data and calculating the difference between the actual energy consumtion and the predicted values, potential anomalies can be detected. The final plot visualizes the actual energy usage values (green dots) and displays a shaded region(orange) representing the predicted interval. Observations outside this interval suggest anomalies or irregularities in the energy usage data. From a technical point of view, they deviate significantly from the predicted values, indicating potential unusual patterns or errors in the energy usage data. From a common sense perspective, these observations outside the predicted interval can be considered anomalies as they represent energy consumption that falls outside the expected range, suggesting abnormal or unexpected behavior in the energy usage patterns, which may require further investigation.

### Seasonal-Trend Decomposition(STD)

```python
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

df_std = df.reset_index(drop = True)
df_std['Datetime'] = pd.to_datetime(df_std['Datetime'])
df_std.index = pd.to_datetime(df_std['Datetime'])

# Decompose data to find trend seasonal and residual
decomposition = seasonal_decompose(df_std['DOM_MW'], model = 'additive')

# Extract decomposed trends
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

decomposition.plot()
plt.show()
```

```python
# Check the trend for the first 300 days

decomposition.trend[1:300].plot()
```

```python
# Check the seasonal for the first 300 days

decomposition.seasonal[1:300].plot()
```

```python
# Plot a graph that contain both residual and original data to detect the potential anomalies
plt.figure(figsize=(10, 6))
plt.plot(df_std['Datetime'], residual, label='Residual')
plt.plot(df_std['Datetime'], df_std['DOM_MW'], c='orange', label='Original Data')

plt.xlabel('Date')
plt.ylabel('DOM_MW')
plt.legend()
plt.show()
```

### Conclusion
#### By separating the data into trend, seasonal, and residual components, STD allows for a clearer understanding of underlying patterns and variations. The plot above, displaying the original data(orange shaded region) alongside the residual component(blue shaded region), facilitates visual inspection and comparison. Anomalies can be detected by scrutinizing deviations between the original data and the residual component. From a technical point of view,unexpected spikes, patterns, or outliers in the residuals can signify anomalies, because they manifest as significant deviations between the original data and the residual component, indicating unexpected variations beyond the seasonal and trend patterns captured by the model. From a common sense perspective, these outliers detected in the residuals can be considered anomalies as they represent abnormal patterns or extreme values that stand out from the expected behavior of the time series, highlighting potential irregularities or exceptional circumstances that merit further investigation.

```python
from prophet import Prophet

# Convert original time and energy demand columns to 'ds' and 'y' to fulfill prophet's requirement
df_prophet = df
df_prophet.columns = ['ds', 'y']
df_prophet = df_prophet.reset_index(drop = 'index')

# Convert ds(Datetime) to datetime class
df_prophet.index = pd.to_datetime(df_prophet['ds'])

# As The data had been converted to daily, so I used a dialy model here
demand_model = Prophet(daily_seasonality=True)
demand_model.fit(df_prophet)
```

```python
from prophet.plot import add_changepoints_to_plot

# Set a future of one year, and use the prophet model to predict future energy need
future= demand_model.make_future_dataframe(periods=365, freq='d')
forecast = demand_model.predict(future)

# Plot the historucal data trends, change points, detected anomalies and energy demand forcasting
fig = demand_model.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), demand_model, forecast)
```

### Conclusion
#### The Prophet model incorporates components such as trend, seasonality, and holiday effects to establish a baseline of expected behavior, allowing for the detection of anomalies. From technical point of view, anomalies are identified as data points that deviate significantly from established patterns, indicating irregularities or unexpected events. From a common sense perspective, the Prophet model's ability to generate future forecasts aids in identifying anomalies in energy consumption. Anomalies manifest as observations that fall outside the tolerance range or exhibit notable deviations from predicted trends, signaling abnormal energy usage patterns that defy historical expectations. Additionally, the model's inclusion of changepoints reveals significant shifts or disruptions in energy consumption, providing insights into anomalies and potential forecasting challenges. Visualization of forecasted values alongside historical data facilitates the identification of anomalies, prompting further investigation into underlying causes or contributing factors.
